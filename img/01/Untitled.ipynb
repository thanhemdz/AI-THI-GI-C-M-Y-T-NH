```python
import matplotlib.pyplot as plt

# Dữ liệu từ bảng
categories = ["Xuất sắc", "Tốt", "Trung bình", "Tệ", "Quá tệ"]
articles = ["Bài viết 1", "Bài viết 2", "Bài viết 3", "Bài viết 4", "Bài viết 5", "Bài viết 6", "Bài viết 7", "Bài viết 8", "Bài viết 9", "Bài viết 10"]
data = [
    [62, 30, 8, 0, 0],
    [56, 25, 18, 1, 0],
    [64, 19, 17, 0, 0],
    [54, 26, 18, 2, 0],
    [60, 31, 9, 0, 0],
    [78, 21, 1, 0, 0],
    [61, 34, 5, 0, 0],
    [79, 16, 5, 0, 0],
    [68, 22, 9, 1, 0],
    [75, 21, 4, 0, 0]
]

# Chuyển vị ma trận dữ liệu để dễ dàng vẽ
data_transposed = list(map(list, zip(*data)))

# Màu cho các cột
colors = ['#4CAF50', '#2196F3', '#FFC107', '#FF5722', '#E91E63']

# Vẽ biểu đồ
fig, ax = plt.subplots(figsize=(6, 6))
width = 0.15
x = range(len(articles))

for i in range(len(categories)):
    ax.bar(
        [pos + width * i for pos in x],
        data_transposed[i],
        width=width,
        label=categories[i],
        color=colors[i]
    )

ax.set_xlabel('Bài viết')
ax.set_ylabel('Phần trăm')
ax.set_xticks([pos + width * 2 for pos in x])
ax.set_xticklabels(articles)
ax.legend(categories, title="Đánh giá")

plt.xticks(rotation=45, ha="right")
plt.tight_layout()
plt.show()
```
